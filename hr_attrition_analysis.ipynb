```python
# =====================================================
# 1. Import Required Libraries
# =====================================================

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt


# =====================================================
# 2. Load Dataset
# =====================================================

data = pd.read_csv(r"data/HR_Employee_Attrition.csv")
print("Dataset loaded successfully!")
print(data.shape)
print(data.head())


# =====================================================
# 3. Initial Data Inspection
# =====================================================

print("\nMissing values:")
print(data.isnull().sum())


# =====================================================
# 4. Data Cleaning & Target Encoding
# =====================================================

if "EmployeeID" in data.columns:
    data = data.drop("EmployeeID", axis=1)

data["Attrition"] = data["Attrition"].map({"Yes":1, "No":0})

X = data.drop("Attrition", axis=1)
y = data["Attrition"]


# =====================================================
# 5. Handling Missing Values
# =====================================================

num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns

for col in num_cols:
    X[col].fillna(X[col].median(), inplace=True)

for col in cat_cols:
    X[col].fillna(X[col].mode()[0], inplace=True)


# =====================================================
# 6. Encoding Categorical Variables
# =====================================================

X = pd.get_dummies(X, columns=cat_cols, drop_first=True)

print("Dataset Shape:", data.shape)
print("\nColumn Names:\n", data.columns.tolist())
print("\n--- Dataset Info ---")
print(data.info())
print("\n--- Summary Statistics ---")
display(data.describe())


# =====================================================
# 7. Exploratory Data Analysis (EDA)
# =====================================================

# Distribution of Target Variable (Attrition)
plt.figure(figsize=(5,4))
attr_counts = data['Attrition'].value_counts()
plt.bar(attr_counts.index, attr_counts.values, color=['skyblue', 'salmon'])
plt.title("Attrition Count (Yes/No)")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()


# =====================================================
# 8. Categorical Variables vs Attrition
# =====================================================

cat_cols = data.select_dtypes(include=['object']).columns
plt.figure(figsize=(15,10))

i = 1
for col in cat_cols[:6]:
    plt.subplot(2,3,i)
    grouped = data.groupby([col,'Attrition']).size().unstack(fill_value=0)
    x = np.arange(len(grouped))
    width = 0.35
    plt.bar(x - width/2, grouped.iloc[:,0], width, label=grouped.columns[0], color='skyblue')
    if len(grouped.columns) > 1:
        plt.bar(x + width/2, grouped.iloc[:,1], width, label=grouped.columns[1], color='salmon')
    plt.title(col + " vs Attrition")
    plt.xticks(x, grouped.index, rotation=45)
    plt.legend()
    i = i + 1

plt.tight_layout()
plt.show()


# =====================================================
# 9. Numerical Feature Distributions
# =====================================================

num_cols = data.select_dtypes(include=['int64', 'float64']).columns
data[num_cols].hist(bins=20, figsize=(15, 10), edgecolor='black', color='skyblue')
plt.suptitle("Distribution of Numerical Variables", size=16)
plt.show()


# =====================================================
# 10. Correlation Analysis
# =====================================================

if 'Attrition' in data.columns:
    if data['Attrition'].dtype == 'object':
        data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})
    corr = data.corr(numeric_only=True)
    corr_with_target = corr['Attrition'].sort_values(ascending=False)
    print("\n--- Correlation of Features with Attrition ---")
    print(corr_with_target)


# =====================================================
# 11. Train-Test Split
# =====================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)


# =====================================================
# 12. Feature Scaling
# =====================================================

num_cols = X_train.select_dtypes(include=["int64", "float64"]).columns
scaler = StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])


# =====================================================
# 13. Logistic Regression Model
# =====================================================

lr = LogisticRegression(max_iter=1000, solver='liblinear')
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
y_prob_lr = lr.predict_proba(X_test)[:,1]

print("\n--- Logistic Regression Results ---")
print(classification_report(y_test, y_pred_lr))
print("ROC AUC:", roc_auc_score(y_test, y_prob_lr))

coeff = pd.DataFrame({"Feature": X_train.columns, "Coef": lr.coef_[0]})
coeff["Odds"] = np.exp(coeff["Coef"])
print("\nTop 10 Logistic Regression features:")
print(coeff.sort_values("Coef", key=abs, ascending=False).head(10))


# =====================================================
# 14. Random Forest Model
# =====================================================

rf = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]

print("\n--- Random Forest Results ---")
print(classification_report(y_test, y_pred_rf))
print("ROC AUC:", roc_auc_score(y_test, y_prob_rf))

importances = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": rf.feature_importances_
}).sort_values("Importance", ascending=False)

print("\nTop 10 Important Features (Random Forest):")
print(importances.head(10))


# =====================================================
# 15. Confusion Matrix Visualization
# =====================================================

cm_lr = confusion_matrix(y_test, y_pred_lr)
cm_rf = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(cm_lr, cmap="Blues")
plt.title("Logistic Regression CM")
plt.xlabel("Predicted")
plt.ylabel("Actual")
for (i, j), val in np.ndenumerate(cm_lr):
    plt.text(j, i, val, ha='center', va='center')

plt.subplot(1,2,2)
plt.imshow(cm_rf, cmap="Greens")
plt.title("Random Forest CM")
plt.xlabel("Predicted")
plt.ylabel("Actual")
for (i, j), val in np.ndenumerate(cm_rf):
    plt.text(j, i, val, ha='center', va='center')

plt.tight_layout()
plt.show()


# =====================================================
# 16. ROC Curve Comparison
# =====================================================

fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

plt.plot(fpr_lr, tpr_lr, label=f"Logistic Regression (AUC={roc_auc_score(y_test, y_prob_lr):.2f})")
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC={roc_auc_score(y_test, y_prob_rf):.2f})")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

print("\nDone! Both models trained successfully :)")


# =====================================================
# 17. Identifying Employees Most Likely to Leave
# =====================================================

pred_prob = rf.predict_proba(X_test)[:,1]
results = X_test.copy()
results["Actual_Attrition"] = y_test.values
results["Attrition_Probability"] = pred_prob

results_sorted = results.sort_values("Attrition_Probability", ascending=False)

print("\nTop 10 Employees Most Likely to Leave:")
print(results_sorted.head(10)[["Attrition_Probability"]])


# =====================================================
# 18. Why Employees Are Leaving (Interpretability)
# =====================================================

coeff = pd.DataFrame({
    "Feature": X_train.columns,
    "Coefficient": lr.coef_[0]
})
coeff["Odds_Ratio"] = np.exp(coeff["Coefficient"])

coeff = coeff.sort_values("Coefficient", key=abs, ascending=False)

print("\nTop Reasons Why Employees Leave (based on Logistic Regression):")
print(coeff.head(10))

print("\nMain Takeaways:")
print("- Positive coefficient = increases chance of leaving")
print("- Negative coefficient = decreases chance of leaving")
print("- So higher 'OverTime' or lower 'JobSatisfaction' probably cause attrition.")
```
